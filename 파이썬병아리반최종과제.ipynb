#시험준비 및 점검을 위한 O/X 퀴즈

## 단원별 점검이 가능한 OX퀴즈로 즐겁게 시험준비!

####로그인 후 단원을 선택하면 시작됩니다.
####즐공 즐겜!

```python
import csv
import random
import time


subject = ['과학', '수학', '영어', '국어']   #문제파일이 준비된 과목 입력

print('-'*50)                                #오프닝
print('{0:' '^50}'.format('안녕하세요!!'))
print('{0:' '^50}'.format('즐겁게 퀴즈로 점검해봅시다!'))
print('-'*50)
time.sleep(1)                                #잠시 멈추기

input_subject = ''                            #while함수로 들어가기 전 변수 미리 만들어두기
input_grade = 0                               #while함수로 들어가기 전 변수 미리 만들어두기

while True :                                  #준비된 과목만 입력될 수 있도록 반복문 사용
  print('어떤 과목의 문제를 풀어볼까요?' + '\n' + '(현재 선택 가능한 과목:' + str(subject))
  input_subject = input('과목선택: ')
  if input_subject in subject :
    print('학년(1~3)을 입력해주세요')
    break                                     #목적한 과목이 리스트에 있으면 break를 통해 탈출
  else :
    print('잘못 입력하셨습니다.')             #다시 입력할 수 있도록 안내하고 입력부터 반복

while True :
  input_grade = input('학년선택: ')
  if input_grade == 1 or 2 or 3 :             #학년은 1~3 뿐이므로 입력되면 break를 통해 탈출
    print('알겠습니다! 이제 시작해볼까요?')
    break
  else :
    print('잘못 입력하셨습니다.')             #다시 입력할 수 있도록 안내하고 입력부터 반복


if input_subject == subject[0] :              #입력된 정보를 이용하여 문제파일을 찾아 정보 읽기
  if int(input_grade) == 1 :                       #미리 모든 파일을 읽어두어도 되나, 양이 많아지면 읽는 속도가 느려질 것을 대비하여 문제파일 하나만 읽어들이기
    data = csv.reader(open('OXQ_middle1_science.csv'))   #만약 모든 파일을 읽어들여놓는다면 각각을 data1등으로 다르게 읽어서 자료처리
  elif int(input_grade) == 2 :
    data = csv.reader(open('OXQ_middle2_science.csv', encoding = 'cp949'))
  else :
    data = csv.reader(open('OXQ_middle3_science.csv', encoding = 'cp949'))

elif input_subject == subject[1] :
  if int(input_grade) == 1 :
    data = csv.reader(open('OXQ_middle1_math.csv', encoding = 'cp949'))  
  elif int(input_grade) == 2 :
    data = csv.reader(open('OXQ_middle2_math.csv', encoding = 'cp949'))
  else :
    data = csv.reader(open('OXQ_middle3_math.csv', encoding = 'cp949'))

elif input_subject == subject[2] :
  if int(input_grade) == 1 :
    data = csv.reader(open('OXQ_middle1_english.csv', encoding = 'cp949'))  
  elif int(input_grade) == 2 :
    data = csv.reader(open('OXQ_middle2_english.csv', encoding = 'cp949'))
  else :
    data = csv.reader(open('OXQ_middle3_english.csv', encoding = 'cp949'))

elif input_subject == subject[3] :
  if int(input_grade) == 1 :
    data = csv.reader(open('OXQ_middle1_korean.csv', encoding = 'cp949'))  
  elif int(input_grade) == 2 :
    data = csv.reader(open('OXQ_middle2_korean.csv', encoding = 'cp949'))
  else :
    data = csv.reader(open('OXQ_middle3_korean.csv', encoding = 'cp949'))


next(data)                             #첫번째줄(단원) 제외
next(data)                             #두번째줄(항목) 제외
data = list(data)                      #반복하여 문제를 풀 때를 대비하여 자료를 리스트로 변환


question = []                          #문제를 입력할 리스트 작성
answer = []                            #답을 입력할 리스트 작성
score = []                             #점수를 입력할 리스트 작성


time.sleep(1)
input_unit = input('문제가 출제 될 단원을 정해주세요 :')   #문제파일 내의 단원에 따라 문제 구분
                                                           #단원별 문제,답,점수 열이 입력되어 있음.

for row in data :
  question.append(row[3*int(input_unit) - 3])             #원하는 단원이 선택되도록 열 번호 3*input_unit - 3 입력                            
  answer.append(row[3*int(input_unit) - 2])               #문제 열의 우측이 답이므로 3*input_unit - 2 입력
  score.append(row[3*int(input_unit) - 1])                    #답 열의 우측이 답이므로 3*input_unit - 1 입력

print(f'현재 풀 수 있는 문제 수는 {len(question)}개 입니다.')   #파일에 입력되어 있는 문제의 종류 수를 알려줌
input_try = input('문제를 몇 번 푸시겠습니까? :')    #문제 풀 수 입력
time.sleep(1)                                        #잠시 기다리기
print('='*50)
print('{0:=^48}'.format('준비'))
print('='*50)
time.sleep(1)
print('{0:=^48}'.format('시작!!'))

random_num = []                                     #문제가 중복되지 않도록 랜덤변수를 생성할 리스트 작성

while len(random_num) < int(input_try) :            #랜덤 리스트에 문제푸는 횟수가 될 때까지 반복입력 
  k = random.randint(1,len(question))               #숫자를 랜덤으로 생성(랜덤의 범위는 문제가 들어있는 리스트의 수만큼)
  if k not in random_num :                          #랜덤 리스트에 중복된 수가 없다면
    random_num.append(k)                            #지금 숫자를 리스트에 넣기

score_my = 0                                        #현재 점수를 출력하기 위한 변수
score_all = 0                                       #만점을 출력하기 위한 변수


for i in random_num :                               #랜덤 리스트에 있는 숫자대로 문제 출제
  print(f'{score[i]}점 문제입니다.')                #자료에 입력된 문제의 점수를 표시
  time.sleep(1)
  inputOX = input(question[i])                      #질문을 프린트하고 답을 입력받음
  if inputOX == answer[i] :                         #입력된 답을 파일에 저장된 답과 비교
    print('정답입니다')                             #정답일 경우 정답 메시지 출력하고
    score_my = score_my + int(score[i])             #자신의 점수를 누적함
    score_all = score_all + int(score[i])           #전체 점수를 동일하게 누적하여, 만점 계산
  else:
    print('틀렸습니다')                             #답이 틀린 경우 오답 메시지 출력
    score_all = score_all + int(score[i])           #내 점수가 아닌 전체점수만 누적하여 만점 계산

print(f'당신의 점수는 {score_my}/{score_all}점입니다!!')   #내가 받은 점수와 만점점수를 이용하여 성적 표시



#로그인(아이디,비밀번호)
#점수를 서버에 등록하여 학교 또는 지역별 비교
#문제 수를 정해놓고 하는 버전 + 정해진 시간 안에 최대한 문제를 풀어보는 경쟁모드
#점수를 지속적으로 누적 저장해서 점수변화에 대한 그래프(성적 변화 그래프)
```
